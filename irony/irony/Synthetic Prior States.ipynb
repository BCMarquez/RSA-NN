```python
import numpy as np
prior_states = {}
for i in range(100):
    a = np.random.random(5)
    a /= a.sum()
    prior_states["WC"+str(i+1)] = {state+1:vals for state, vals in enumerate(a.tolist())}

with open('synthetic_prior_states.json', 'w') as fp:
    fp.write(str(prior_states))
```
